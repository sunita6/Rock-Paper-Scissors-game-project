```python
#random able to use the different tools inside to randomize the computer’s actions in the game. 
#It works by assigning each attribute within it to a value you specify.


import pyttsx3
from random import randint

computer_score=0
player_score=0
g=0

a = ["rock", "paper", "scissors"]
```

```python
# function for text to speech
engine=pyttsx3.init()
def texttospeech(txt):
    pyttsx3.init()
    voices=engine.getProperty("voices")
    engine.setProperty("voice",voices[g].id)
    rate=engine.getProperty("rate")
    engine.setProperty("rate",145)
    engine.say(txt)
    engine.runAndWait()
```

```python
# code for get_user_selection(), which doesn’t take in any arguments and returns an Action:

def get_user_selection():
    texttospeech("Enter a choice zero for rock,1 for paper,2 for scissor")
    user_action = int(input("Enter a choice (rock[0], paper[1], scissors[2]): "))
    print("Player choose :",a[user_action])
    texttospeech("player choose "+a[user_action] )
    return user_action
```

```python
#the values for Action range from 0 to 2
#random.randint() returns a random value between a specified minimum and maximum (inclusive).

import random
def get_computer_selection():
    selection = random.randint(0,2) 

    #assign what the computer chose to rock, paper, or scissors 
    if selection == 0: 
        print("computer choose :",a[0])
        texttospeech("computer choose "+a[0])
    elif selection == 1: 
        print("computer choose :",a[1])
        texttospeech("computer choose "+a[1])
    else: 
        print("computer choose :",a[2])
        texttospeech("computer choose "+a[2])

    #return value 
    return selection 
```

```python
#Now determine_winner function will take two arguments, the user’s action and the computer’s action. 
#It doesn’t need to return anything since it’ll just display the result to the console:

def determine_winner(user_action, computer_action):
    global computer_score
    global player_score
    if str(user_action) == str(computer_action):
        print("Both players selected the same ,It's a tie!")
        texttospeech("Both players selected the same It's a tie!")
    elif user_action == 0:
        if computer_action == 2:
            player_score +=1
            print("Rock smashes scissors! You win!")
            texttospeech("Rock smashes scissors! You win!")
        else:
            computer_score +=1
            print("Paper covers rock! You lose.")
            texttospeech("Paper covers rock! You lose.")
    elif user_action == 1:
        if computer_action == 0:
            player_score +=1
            print("Paper covers rock! You win!")
            texttospeech("Paper covers rock! You win!")
        else:
            computer_score +=1
            print("Scissors cuts paper! You lose.")
            texttospeech("Scissors cuts paper! You lose.")
    elif user_action == 2:
        if computer_action == 1:
            player_score +=1
            print("Scissors cuts paper! You win!")
            texttospeech("Scissors cuts paper! You win!")
        else:
            computer_score +=1
            print("Rock smashes scissors! You lose.")
            texttospeech("Rock smashes scissors! You lose.")
```
